```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = [10,20,30,40,50,60,70,80,90]
a=[1,2,3,4,5,6,7,8,9]
b=[1,2,3,4,5,6,7,8,9]

plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('x vs a&b')
plt.xlabel('x')
plt.ylabel('a&b')
plt.show()
```
